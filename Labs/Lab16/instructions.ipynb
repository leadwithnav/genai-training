# JMeter MCP Server Setup: Step-by-Step Instructions

Follow these steps to set up and run the JMeter MCP Server for distributed JMeter test execution via MCP:

---

## 1. Install Required Dependencies

Open a terminal and run:
```bash
pip install numpy matplotlib uv
```

---

## 2. Clone the JMeter MCP Server Repository

```bash
git clone https://github.com/QAInsights/jmeter-mcp-server.git
cd jmeter-mcp-server
```

---

## 3. Install Python Dependencies with uv

```bash
uv sync
uv pip install python-dotenv numpy matplotlib
```

---

## 4. Create a `.env` File

In the `jmeter-mcp-server` directory, create a file named `.env` with the following content (update the paths as needed):

```
JMETER_HOME=C:\Tools\apache-jmeter-5.6.3
JMETER_BIN=C:\Tools\apache-jmeter-5.6.3\bin\jmeter.bat
JMETER_JAVA_OPTS=-Xms1g -Xmx2g
```

---

## 5. Start the JMeter MCP Server

```bash
uv run jmeter_server.py
```

---

## 6. Add JMeter MCP Server to MCP Config

In your `mcp.json` configuration, add the following entry under `servers`:

```json
"jmeter": {
    "command": "C:\\Users\\Lenovo\\AppData\\Local\\Programs\\Python\\Python313\\Scripts\\uv.exe",
    "args": [
        "--directory",
        "D:\\jmeter-mcp-server",
        "run",
        "jmeter_server.py"
    ]
}
```

- Adjust the paths if your installation directories are different.

---

## 7. Verify MCP Integration

- Restart your MCP environment if needed.
- The JMeter MCP server should now be available as a tool in your MCP UI or workflow.
- You can now execute JMeter tests via the MCP protocol.

---

**Tip:**
- Ensure Java and JMeter are installed and the paths in `.env` are correct.
- For troubleshooting, check the server logs and MCP UI for errors.
