# Lab: Using GitHub Copilot for Test Plan Generation

Objective: Learn how to use GitHub Copilot, specifically the Copilot Chat feature (Ask mode), to generate a test plan for a software feature.

## 1. Introduction to GitHub Copilot
- GitHub Copilot is an AI-powered coding assistant that helps you write code, generate documentation, and automate tasks directly in your editor.
- It can suggest code, explain code, and answer questions using natural language prompts.

## 2. Copilot Usage
- **Inline Suggestions**: Copilot suggests code as you type.
- **Copilot Chat**: Use the chat sidebar or command palette to ask questions, get explanations, or generate code/tests.

## 3. Copilot Chat Modes
- **Ask Mode**: Provides answers or code snippets based on your questions.
- **Edit Mode**: Allows you to edit generated content iteratively.
- **Plan Mode**: Helps you create structured plans or outlines.
- **Agent Mode**: Enables you to create and manage AI agents for automated tasks.

> This lab is focused only on using GitHub Copilot features. Prompt engineering techniques are not covered here.

# Part 2 — Core Copilot Features Every Tester Needs

## 2.1 Inline Suggestions — Testers' Version

Copilot's inline suggestions aren't just for developers. As a tester, you can use them to:

- **Write test scenarios** in Markdown or plain text
- **Complete test case tables** 
- **Draft test data** in structured formats (CSV, JSON, SQL)

### ✏️ Try It Now

<span style="color:orange">Step 1:</span> Open a **new file** in VS Code: <kbd>Ctrl</kbd>+<kbd>N</kbd>  
<span style="color:orange">Step 2:</span> Save it as `my-test-scenarios.md` (<kbd>Ctrl</kbd>+<kbd>S</kbd>)  
<span style="color:orange">Step 3:</span> Type the following and then press **Enter**:

```
## Test Scenarios for Add to Cart

| ID | Scenario | Expected Result |
|----|-
```

> 💡 Copilot recognises the Markdown table structure and will suggest completing the table with test scenarios. Press **Tab** to accept.

---

## 2.2 Copilot Chat — Ask Mode

Use **Ask mode** when you want Copilot to answer questions without touching your files.

### ✏️ Try It Now

<span style="color:orange">Step 4:</span> Open the **Copilot Chat panel** (speech-bubble icon in the sidebar)  
<span style="color:orange">Step 5:</span> Make sure the mode selector shows **Ask**  
<span style="color:orange">Step 6:</span> Paste this prompt and press Enter:

```
Explain what a good test scenario for an "Add to Cart" feature should cover.
Give examples of positive cases, negative cases, and edge cases.
```

> 💡 Copilot will respond with a structured explanation. This is useful for **learning** and for **reviewing** whether your test scenarios are complete.

---

## 2.3 Inline Chat — Quick Edits at the Cursor

### ✏️ Try It Now

<span style="color:orange">Step 7:</span> In your `my-test-scenarios.md` file, put your cursor at the end of the file  
<span style="color:orange">Step 8:</span> Press <kbd>Ctrl</kbd>+<kbd>I</kbd>

---

## 2.4 Using `#file` to Give Copilot Context

Copilot is smarter when you point it at the right files.

### ✏️ Try It Now

<span style="color:orange">Step 11:</span> In the Copilot Chat panel, type:

```
#file:ecommerce-Jira.md

Based on the story EC-201 in this file, what test cases should I write?
```

> 💡 The `#file:` prefix tells Copilot exactly which file to read. This is called **context attachment**. Without it, Copilot guesses from your open tabs — with it, you get much more accurate results.

Other useful context prefixes:

| Prefix | What it attaches |
|--------|------------------|
| `#file:filename.md` | A specific file |
| `#selection` | Your currently selected text |
| `#codebase` | Searches the whole project (Agent mode) |

## 3.1 The Source: Jira Story EC-302 — Add Funds to Wallet

Here is the full story you will work with:

---

**EC-302 — Add Funds to Wallet**  
**Type:** Story | **Priority:** High | **Points:** 2

> *As a user, I want to add money to my wallet, so that I can make purchases in the store.*

**Acceptance Criteria:**
- \[AC1\] Wallet page has an input field for amount and an "Add Funds" button.
- \[AC2\] Clicking "Add Funds" with a valid amount updates the balance.
- \[AC3\] Balance updates immediately on the page after adding funds.
- \[AC4\] Transaction is recorded in `wallet_transactions` table.
- \[AC5\] Negative or zero amounts are rejected with a validation message.
- \[AC6\] Maximum single top-up is $10,000.

**API:** `POST /api/wallet/add`  
**Body:** `{ sessionId, amount }`

---

**Out of Scope (from the project):**
- Real payment gateway — no credit card processing
- Email notifications
- Multi-currency support

## 3.2 Exercise A — Generate Initial Test Scenarios

<span style="color:orange; font-size:1.1em">🟢 Your First Prompt</span>

<span style="color:orange">Step 1:</span> Open the **Copilot Chat panel** → **Ask mode**

<span style="color:orange">Step 2:</span> Paste the prompt below and press Enter:

---

```
You are a QA engineer. Based on the following Jira story, generate manual test scenarios.

Story: EC-302 — Add Funds to Wallet
As a user, I want to add money to my wallet so that I can make purchases.

Acceptance Criteria:
- AC1: Wallet page has an amount input and an 'Add Funds' button.
- AC2: Adding a valid amount updates the balance.
- AC3: Balance updates immediately on the page.
- AC4: Transaction is recorded in the database.
- AC5: Negative or zero amounts are rejected with a validation message.
- AC6: Maximum single top-up is $10,000.

Out of scope: real payments, email notifications, multi-currency.

Generate scenarios in this format:

| ID | Priority | Type | Scenario Title | Precondition | Steps | Expected Result |

Include:
- Positive cases (P0)
- Negative cases (P1)
- Edge cases (P2)

Do not include out-of-scope scenarios.
```

---

> 💡 Wait for Copilot to respond. It should produce a table with 8–15 test scenarios.

## 3.3 What Good Output Looks Like

Here's what a **well-structured** Copilot response for EC-302 should look like:

| ID | Priority | Type | Scenario Title | Precondition | Steps | Expected Result |
|----|----------|------|---------------|-------------|-------|-----------------|
| TC-001 | P0 | Positive | Add valid amount to wallet | Wallet page open, balance is $0 | 1. Enter `100` in amount field 2. Click "Add Funds" | Balance shows $100.00; success message displayed |
| TC-002 | P0 | Positive | Balance updates immediately on page | Wallet page shows $50 | 1. Enter `50` 2. Click "Add Funds" | Balance updates to $100 without page reload |
| TC-003 | P0 | Positive | Transaction recorded in database | User has existing wallet | 1. Add $200 2. Check transaction history | New entry appears in transaction list with type "deposit" |
| TC-004 | P1 | Negative | Reject zero amount | Wallet page open | 1. Enter `0` 2. Click "Add Funds" | Error message: "Amount must be greater than zero" |
| TC-005 | P1 | Negative | Reject negative amount | Wallet page open | 1. Enter `-50` 2. Click "Add Funds" | Error message shown; balance unchanged |
| TC-006 | P1 | Negative | Reject empty amount field | Wallet page open | 1. Leave field blank 2. Click "Add Funds" | Error message: "Amount is required" |
| TC-007 | P2 | Edge | Add exactly $10,000 (maximum) | Wallet page open | 1. Enter `10000` 2. Click "Add Funds" | Balance increases by $10,000 successfully |
| TC-008 | P2 | Edge | Reject amount above $10,000 | Wallet page open | 1. Enter `10001` 2. Click "Add Funds" | Error message: "Maximum top-up is $10,000" |

## 3.4 Exercise B — Spot the Common Mistakes

<span style="color:orange; font-size:1.1em">🔎 Review and Critique</span>

Review the output Copilot gave you. Check for these **common AI test generation mistakes**:

---

### ❌ Mistake 1: Out-of-Scope Scenarios

**Example of a bad scenario Copilot might generate:**

| TC-011 | P1 | Negative | Add funds using an expired credit card | ... | ... | Payment declined message shown |

❌ **This is out of scope** — there is no credit card in this system. Mark it and delete it.

---

### ❌ Mistake 2: Vague Expected Results

**Bad:**
> Expected Result: "Balance is updated correctly."

**Good:**
> Expected Result: "Balance shows $150.00 (was $50.00 + $100.00 added). Success toast message 'Funds added to wallet' is displayed."

---

### ❌ Mistake 3: Missing Preconditions

**Bad:**
> Precondition: "User is logged in."

❌ **This system has no login** — session IDs are used. A correct precondition is:

> Precondition: "Wallet page is open; session ID is `test-123`; current balance is $0.00."

---

### ❌ Mistake 4: Duplicate Scenarios with Different Names

Copilot sometimes generates "Add negative amount" and "Enter amount less than zero" as two separate test cases. They test the same thing.

**Fix:** Merge them or remove the duplicate.

---

### ❌ Mistake 5: Wrong Priority Tags

Critical business rules (like "zero amount rejected") should be **P0 or P1**, not P2.  
UI styling issues and nice-to-haves should be P2.

**Check:** Does Copilot's priority tagging match your team's risk assessment?

## 3.5 Exercise C — Refine with Follow-Up Prompts

<span style="color:orange; font-size:1.1em">🔄 Iterating with Copilot</span>

Good prompting is a **conversation**, not a one-shot command. Use follow-up prompts to improve the output.

---

### Refinement Prompt 1: Fix vague expected results

```
For each scenario you generated, make the 'Expected Result' more specific.
Include the exact balance value, the exact error message text, and whether
the page refreshes or updates in-place.
```

---

### Refinement Prompt 2: Remove out-of-scope cases

```
Remove any test scenarios that involve:
- Credit cards or real payment systems
- Email notifications
- User login or authentication
- Currency conversion

This system uses session IDs, not user accounts.
```

---

### Refinement Prompt 3: Add missing boundary cases

```
I noticed boundary value analysis is incomplete. Add test scenarios for:
- Amount = $0.01 (minimum valid, one cent)
- Amount = $9,999.99 (just under maximum)
- Amount = $10,000 (exactly at maximum)
- Amount = $10,000.01 (just over maximum)
- Amount with more than 2 decimal places (e.g., $10.999)
```

---

### Refinement Prompt 4: Verify AC coverage

```
Review the test scenarios against these acceptance criteria:
AC1, AC2, AC3, AC4, AC5, AC6.

For each AC, tell me which test scenario(s) cover it.
If any AC has no coverage, add a scenario for it.
```

> 💡 This is called **requirement traceability** — ensuring every acceptance criterion has at least one test.

**EC-402 — Cancel a Placed Order**  
**Priority:** High

> *As a user, I want to cancel an order that hasn't been delivered yet, so that I can get a refund if I change my mind.*

**Acceptance Criteria:**

- Clicking "Cancel" shows a confirmation prompt

- Refund is automatically added to the user's wallet

**Order Status Flow:** `placed` → `cancelled` *(only valid transition for cancel)*

### 🔑 CARET Framework — Build Your Prompt

| Letter | Stands for | For EC-402, fill in... |
|--------|------------|------------------------|
| **C** | **Context** | *You are a QA engineer testing an ecommerce order management system...* |
| **A** | **Action** | *Generate manual test scenarios for...* |
| **R** | **Result** | *...covering positive, negative, and edge cases* |
| **E** | **Edge cases** | *...including status transition boundaries (e.g., cancel a delivered order)* |
| **T** | **Table format** | *Format as a Markdown table with columns: ID, Priority, Type, Title, Precondition, Steps, Expected Result* |

### ✅ Positive Cases (P0)

- Email refund notifications
- Partial cancellations (refund only one item)

- Third-party payment gateway interactions

## Key Principles to Remember

<div style="background:#0d1117; border: 1px solid #30363d; padding:18px; border-radius:8px; color:#e6edf3;">

1. **Garbage in, garbage out.** Copilot's output quality depends directly on prompt quality.  
   → More context = better test scenarios.

2. **Always review for scope creep.** AI models don't always respect "out of scope" — check every scenario.

3. **Traceability matters.** Every acceptance criterion should map to at least one test scenario.

4. **Iterate, don't regenerate.** One good follow-up prompt beats starting over from scratch.

5. **You are the expert.** Copilot doesn't know your system, your users, or your team's risk appetite — you do.
