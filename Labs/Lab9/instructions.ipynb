### 1. Install openpyxl

`openpyxl` is the Python library used to create and write `.xlsx` Excel files from scratch.

Run in your terminal:

```bash
pip install openpyxl
```

---

### 2. Set Up the Postman MCP Server

**a) Generate a Postman API Key:**
1. Log in to [postman.com](https://postman.com)
2. Go to **Settings → API Keys**
3. Click **Generate API Key**, give it a name, and copy the key

**b) Set it as an environment variable:**

```powershell
# Windows PowerShell
$env:POSTMAN_API_KEY = "your-api-key-here"
```

**c) Add the Postman MCP Server to your `mcp.json`:**

```json
"com.postman/postman-mcp-server": {
    "type": "stdio",
    "command": "npx",
    "args": [
        "@postman/postman-mcp-server@2.7.0"
    ],
    "env": {
        "POSTMAN_API_KEY": "${input:POSTMAN_API_KEY}"
    },
    "gallery": "https://api.mcp.github.com",
    "version": "2.7.0"
}
```

> 💡 After saving `mcp.json`, reload the window and verify the Postman MCP Server appears in Copilot Agent Mode's tool list.
