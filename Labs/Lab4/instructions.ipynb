# Lab 4: Excel MCP — Read, Execute & Update Test Cases with Playwright

### Introduction

In this Lab, you will use the **Excel MCP Server** and **Playwright MCP Server** together inside GitHub Copilot Agent mode. You will follow a **6-step prompt workflow** to:

1. Read a test case from an Excel sheet and extract its steps, test data, and expected results
2. Execute the test case against the live app using Playwright (UI only)
3. Capture stable selectors and minimal POM notes from the execution

4. Create or update a Playwright script based on the findings
5. Update the Excel sheet with the execution status, actual result, and date
6. Run the test on Chromium and debug until it passes

## Excel MCP Server + Playwright Integration

You will work through **6 prompts** in Copilot Agent mode. Each prompt builds on the previous one — from reading a test case out of Excel, to executing it in the browser, capturing selectors, writing a script, running it until it passes, and finally writing the result back to Excel.

> ⚠️ **Before you start:** Make sure the ecommerce app is running (`docker-compose up` in the `ecommerce/` folder) and GitHub Copilot is in **Agent Mode** with both the **Excel MCP Server** and **Playwright MCP Server** enabled.

<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Overview:</span> <span style="color:#1976d2;">Excel MCP Server</span>

You can configure **Excel MCP Server**  in your environment (`mcp.json`). It allows GitHub Copilot Agent to read from and write to Excel files directly.

In this lab, you will use it to:
- 📖 **Read** test steps, test data, and expected results for `TC_CART_003` from the Excel sheet
- ✅ **Write back** the execution status, actual result, and execution date after the test runs

---

### 🔮 Test Case in Scope

| Field | Value |
|-------|-------|
| Test Case ID | `TC_CART_003` |
| Target App | `http://localhost:3000` |
| Script File | `playwright/tests/add_to_cart.spec.js` |
| Scope | UI testing only (no DB or API checks) |

### 🔧 Available MCP Tools Used in This Lab

| Tool | MCP Server | Purpose |
|------|-----------|--------|
| Excel read | Excel MCP | Read test steps and expected results |
| Excel write | Excel MCP | Update status, actual result, date |
| Playwright browser | Playwright MCP | Execute test steps in the browser |

<span style="font-size:180%; color:#1976d2;">&#8595;</span>

3. Apply best practices — stable locators, no `waitForTimeout`, clear test structure> 💡 **Tip:** Check the output code and ensure the test name includes `TC_CART_003` for easy filtering when running.

4. If the earlier execution had failures, pre-fix the known issues before writing the script
