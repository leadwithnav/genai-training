<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 5:</span> <span style="color:#1976d2;">Custom TestRail MCP Server Overview</span>

A custom **TestRail MCP Server** that can be configured in your environment (`mcp.json`). This server bridges GitHub Copilot Agent with your TestRail instance, enabling you to:

- 📖 **Read** test cases and test data directly from TestRail
- 🔍 **Search** for test cases by name or pattern
- ✅ **Execute** those test cases using Playwright via the browser

The server is located at: `Labs/Lab5/testrail.py`

---

### 🔧 Available TestRail MCP Tools

| Tool | Description |
|------|-------------|
| `get_test_case` | Fetch full details (title, steps, expected results) for a test case by its ID |
| `search_test_cases` | Search for test cases matching a regex pattern in their title |

---

### Tool Details

**`get_test_case(case_id: int)`**
- Retrieves the test case title, steps, and expected results from TestRail
- Example: `get_test_case(59)` → returns details for test case C59

**`search_test_cases(query_regex: str, project_id: int = 1)`**
- Searches all test cases in a project whose title matches the given regex pattern
- Example: `search_test_cases("checkout")` → lists all test cases with "checkout" in the title
