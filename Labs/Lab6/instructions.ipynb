# Lab 6: TestRail MCP — Automation Coverage Workflow

### Introduction

In this Lab, you will use the **custom TestRail MCP Server** and **Playwright MCP Server** together inside GitHub Copilot Agent mode. You will follow a **6-step prompt workflow** to:

1. Read a test case from TestRail
2. Check whether the test case is already automated in the codebase
3. Compare coverage — step by step — between TestRail and existing automation

4. Perform a dry run in Playwright to validate the scenario and find selectors
5. Update or create the Playwright test based on coverage gaps and dry run findings
6. Run, debug, and confirm the test passes

## TestRail MCP Server Integration — Automation Coverage Workflow

You will work through **6 prompts** in Copilot Agent mode. This workflow goes beyond simply executing a test — it first checks whether automation already exists, compares coverage, and only then updates or creates the test based on real gaps.

> ⚠️ **Before you start:** Make sure the ecommerce app is running (`docker-compose up` in the `ecommerce/` folder) and GitHub Copilot is in **Agent Mode** with both the TestRail MCP Server and Playwright MCP Server enabled.

<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 1:</span> <span style="color:#1976d2;">Prompt 1 — Read the TestRail Case</span>

Open a **new Agent Mode chat** in GitHub Copilot and run the following prompt:

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span></div>

<div style="background-color:black; color:green; padding:10px; border-radius:8px;">

Read TestRail test case C59 and extract:
- title
- preconditions
- test data
- test steps
- expected results

Output the details in a structured format.
</div>

**What to expect:** Copilot will call `get_test_case(59)` using the TestRail MCP server and display the full test case details — title, preconditions, test data, steps, and expected results — in a structured, readable format.

<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 2:</span> <span style="color:#1976d2;">Prompt 2 — Check Automation</span>

Before writing or modifying any code, check whether test case C59 is **already automated** in the codebase.

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span>


<div style="background-color:black; color:green; padding:10px; border-radius:8px;">
Check whether TestRail test case C59 is already automated in the codebase.
Look for existing test names, tags, comments, mappings, or code references for C59.

Output:
- Automated / Not Automated
- matching file name(s) and test name(s), if found
</div>

**What to expect:** Copilot will:
1. Search the codebase for any test file, test name, comment, or tag referencing C59
2. Report whether the test case is **Automated** or **Not Automated**
3. If found, list the matching file name(s) and test name(s)

> 💡 **Tip:** This helps you decide whether to create a new test or update an existing one.

<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 3:</span> <span style="color:#1976d2;">Prompt 3 — Compare Coverage</span>

Now compare the TestRail test steps against the existing automation to identify what is covered, partially covered, or missing.

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span>


<div style="background-color:black; color:green; padding:10px; border-radius:8px;">
Compare TestRail test case C59 with the existing automation.
For each TestRail step, mark it as:
- Fully covered
- Partially covered
- Missing

Output:
- step-by-step coverage summary
- missing steps
- partially covered steps
</div>

**What to expect:** Copilot will:
1. Map each TestRail step for C59 to the existing automation code
2. Classify each step as Fully covered, Partially covered, or Missing
3. Output a step-by-step coverage table and highlight gaps that need to be addressed

> 💡 **Tip:** Review the missing and partially covered steps — these will guide the dry run and the code update in the next steps.


<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 4:</span> <span style="color:#1976d2;">Prompt 4 — Dry Run</span>

Before writing any code, perform a **dry run** in the browser to validate the scenario and identify selectors.

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span>


<div style="background-color:black; color:green; padding:10px; border-radius:8px;">
Using the Playwright MCP server, perform a dry run for TestRail test case C59 against http://localhost:3000.
Do not write code yet.
Execute the flow step by step, validate the scenario, and identify selectors that will help with automation.
Output:
- dry run summary
- selectors found
- issues/blockers, if any
</div>

**What to expect:** Copilot will:
1. Use Playwright MCP to open the browser and execute each C59 step against the live app
2. Map each step to real UI elements and capture their selectors
3. Report a dry run summary, the selectors found, and any blockers encountered

> 💡 **Tip:** Review the selectors Copilot finds — these will be used to update the Playwright test in the next step.


<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 5:</span> <span style="color:#1976d2;">Prompt 5 — Update Code</span>

Based on the coverage gaps and dry run results, ask Copilot to update or create the Playwright automation for C59.

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span>

</div>


<div style="background-color:black; color:green; padding:10px; border-radius:8px;">

Based on the coverage gaps and dry run results for TestRail test case C59, update the Playwright automation.
Requirements:
- if automation exists, update it with missing or partial steps
- if automation does not exist, create a new test
- add/update code in playwright/tests/ecommerce.spec.js
- use Playwright best practices, strong locators, and proper assertions

Output:
- summary of code changes
- final Playwright test code
</div>


**What to expect:** Copilot will:
1. Check if a test for C59 already exists in `ecommerce.spec.js`
2. If it exists — add the missing or partially covered steps identified in Prompt 3
3. If it doesn't exist — create a new test from scratch using the dry run findings
4. Apply Playwright best practices, strong locators, and proper assertions throughout


<div style="background-color:black; padding:10px; border-radius:8px;">
<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 6:</span> <span style="color:#1976d2;">Prompt 6 — Run and Debug</span>

Run the Playwright test and let Copilot debug and fix it until it passes.

<span style="font-size:180%; color:#1976d2;">&#8595;</span>
<span style="color:orange;">Copy and paste this prompt into Copilot Agent chat:</span>

</div>


<div style="background-color:black; color:green; padding:10px; border-radius:8px;">

Run the Playwright test for TestRail test case C59 on Chromium.
If the test fails, debug and fix the script, then re-run until it passes.
Output:
- test execution result
- issues found
- fixes applied
- final confirmation that the test passes
</div>


**What to expect:** Copilot will:
1. Run the Playwright test for C59 on the Chromium browser
2. If it fails, analyze the error, apply fixes to the test script, and re-run
3. Repeat until the test passes successfully
4. Provide a final summary with execution results, issues found, fixes applied, and confirmation of success

> 💡 **Tip:** You can also run the test manually in the terminal:
> ```
> npx playwright test tests/ecommerce.spec.js --grep "C59" --project=chromium --reporter=line
> ```
