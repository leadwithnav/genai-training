# Lab 10: API + Database Testing with Postman MCP & Postgres MCP Server

## Introduction

In this lab, you will extend your API testing workflow to include **database-level verification** using the **Postgres MCP Server** alongside the **Postman MCP Server**.

You will follow a **3-prompt workflow** to:

1. **Set up** the Postgres MCP Server to connect to the running ecommerce database
2. **Generate** 5 structured API + DB test scenarios for the `POST /api/cart` endpoint, verifying both the API response and the database state
3. **Execute** all test cases via Postman MCP Server, validate the database state via Postgres MCP, and report combined results

> ⚠️ **Before you start:** Make sure the ecommerce app is running (`docker-compose up` in the `ecommerce/` folder), GitHub Copilot is in **Agent Mode**, and both the **Postman MCP Server** and **Postgres MCP Server** are enabled.

<span style="font-size:120%; font-weight:bold; color:#1976d2;">&#x1F4DD; Step 14:</span> <span style="color:#1976d2;">Prerequisites — Set Up Postgres MCP Server</span>

Before running the Lab 10 prompts, configure the Postgres MCP Server so Copilot can read the live database state.

---

### 1. Postgres Database Connection Details

The ecommerce database is running on:

```
postgresql://postgres:password@localhost:5432/ecommerce_db
```

---

### 2. Add the Postgres MCP Server to your `mcp.json`

Add the following entry to your `mcp.json` configuration file and reload the window:

```json
{
  "mcpServers": {
    "postgres": {
      "command": "npx",
      "args": [
        "-y",
        "@modelcontextprotocol/server-postgres",
        "postgresql://postgres:password@localhost:5432/ecommerce_db"
      ]
    }
  }
}
```

> 💡 After saving `mcp.json`, reload the VS Code window and verify the **Postgres MCP Server** appears in Copilot Agent Mode's tool list alongside the Postman MCP Server.

---

### 3. Verify the connection

Once enabled, you can ask Copilot Agent to run a quick sanity check:

```
Use the Postgres MCP Server to run: SELECT COUNT(*) FROM cart_items;
```

If it returns a number, your Postgres MCP Server is connected and ready.
